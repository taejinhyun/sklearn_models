## About the dataset: Heart Disease

 심장질환 자료는 총 76개의 속성으로 이루어져 있으나, 공개된 모든실험에서 사용한 14개 속성으로 이루어진 자료를 이용하고자 한다. 최근까지 연구된 Cleveland database(n=303)를 이용하였으며, 총 303명의 환자자료 중 164명이 심장질환이었다.
 
 14개 자료 필드는 아래의 표와 같으며, 측정방법과 검사에 대한 부가적인 설명은 첨부파일로 작성하였다.
 
 변수명 | 설명
 ---|---
age| 환자 나이
sex| 환자성별 (1 = male; 0 = female)
cp| 흉통유형 (1=정형 협심증 통증, 2=비정형 협심증 통증, 3=비정형 협심증 통증, 4=무증상) 
trestbps | 휴식시 혈압(병원 입원시, mm Hg
chol| 총콜레스테롤, mg/dl
fbs| 공복시 혈당, > 120 mg/dl) (1 = true; 0 = false)
restecg| 심전도 결과
thalach| 최대 심박수
exang| 운동형 협심증 여부(1 = yes; 0 = no)
oldpeak| ST depression induced by exercise relative to rest
slope| ST segment의 경사 (1=upsloping, 2=flat, 3=downsloping
ca| 채색된 주요 혈관수 (0-3) 
thal| 결손된 혈관 3 = normal; 6 = fixed defect; 7 = reversable defect
target| 심장질환 여부 1 or 0(target)

```python
#자료 불러오기 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
heartData = pd.read_csv('heart.csv')
```

```python
print('sex =', heartData['sex'].unique())
print('cp =', heartData['cp'].unique())
print('restecg =', heartData['restecg'].unique())
print('exang =', heartData['exang'].unique())
print('slope =', heartData['slope'].unique())
print('ca =', heartData['ca'].unique())
print('thal =', heartData['thal'].unique())
```

```python
# 타겟(심장질환) 환자수 확인
print('number of Heart disease patient =', heartData[(heartData['target'] ==1) ].count()[1])
```

```python
heartData['target'].value_counts().plot(kind='bar')
plt.show()
```

```python
# Correlation test
sns.heatmap(heartData.corr(),annot=True) 
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```
